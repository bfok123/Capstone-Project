```python
from textgenrnn import textgenrnn
import pandas
import numpy as np


data = pandas.read_csv('lyrics_clean.csv')
lyrics = data.lyrics.to_numpy()
```

```python
file = open('lyrics.txt', 'w')
file.write('')
file = open('lyrics.txt', 'a')
for i in range(len(lyrics)):
    lyric = lyrics[i]
    if (type(lyric) == float):
        continue
    lyric = str(lyric.encode('UTF-8'))
    lyric = lyric.replace('\\n', '\n')
    lyric = lyric.replace('"b"', " ")
    file.write(lyric)
```

```python
textgen = textgenrnn()
textgen.train_from_largetext_file('lyrics.txt', num_epochs=1)
```

```python
textgen.generate(5)
```

```python
textgen2 = textgenrnn()
textgen2.train_from_largetext_file('lyrics.txt', 
                                  num_epochs=1,
                                 word_level=True)
```
